```python
df = df.drop('Data_Tahun', axis=1)
```

**Bersihkan Data**

```python
df_cleaned = df.dropna(axis=0)
```

```python
df_cleaned.info()
```

**Ubah Tipe Data Bebebrapa Kolom**

```python
df_cleaned['Nama_Prov'] = df_cleaned['Nama_Prov'].astype('category')
df_cleaned['Nama_Kab'] = df_cleaned['Nama_Kab'].astype('category')
df_cleaned['Nama_Kec'] = df_cleaned['Nama_Kec'].astype('category')
```

```python
df_cleaned['PAD_2023'] = df_cleaned['PAD_2023'].astype(int)
df_cleaned['PAD_2022'] = df_cleaned['PAD_2022'].astype(int)
df_cleaned['DD_2023'] = df_cleaned['DD_2023'].astype(int)
df_cleaned['DD_2022'] = df_cleaned['DD_2022'].astype(int)
df_cleaned['Pajak_dan_Retribusi_2023'] = df_cleaned['Pajak_dan_Retribusi_2023'].astype(int)
df_cleaned['Pajak_dan_Retribusi_2022'] = df_cleaned['Pajak_dan_Retribusi_2022'].astype(int)
df_cleaned['Al_DD_2023'] = df_cleaned['Al_DD_2023'].astype(int)
df_cleaned['Al_DD_2022'] = df_cleaned['Al_DD_2022'].astype(int)
df_cleaned['Bantuan_Prov_2023'] = df_cleaned['Bantuan_Kab_Kot_2023'].astype(int)
df_cleaned['Bantuan_Prov_2022'] = df_cleaned['Bantuan_Kab_Kot_2022'].astype(int)
df_cleaned['Bantuan_Kab_Kot_2023'] = df_cleaned['Bantuan_Kab_Kot_2022'].astype(int)
df_cleaned['Bantuan_Kab_Kot_2022'] = df_cleaned['Bantuan_Kab_Kot_2022'].astype(int)
df_cleaned['Lainnya_2023'] = df_cleaned['Lainnya_2023'].astype(int)
df_cleaned['Lainnya_2022'] = df_cleaned['Lainnya_2022'].astype(int)
```

```python
df_cleaned['Total_APBDES_Tahun_2023'] = df_cleaned['Total_APBDES_Tahun_2023'].astype(int)
df_cleaned['Pbrdayaan_Masyrkt_2022'] = df_cleaned['Pbrdayaan_Masyrkt_2022'].astype(int)
df_cleaned['Pbrdayaan_Masyrkt_2023'] = df_cleaned['Pbrdayaan_Masyrkt_2023'].astype(int)
df_cleaned['Klautn_PerIkanan_2022'] = df_cleaned['Klautn_PerIkanan_2022'].astype(int)
df_cleaned['Klautn_PerIkanan_2023'] = df_cleaned['Klautn_PerIkanan_2023'].astype(int)
df_cleaned['Ptanian_Pternakn_2022'] = df_cleaned['Ptanian_Pternakn_2022'].astype(int)
df_cleaned['Ptanian_Pternakn_2023'] = df_cleaned['Ptanian_Pternakn_2023'].astype(int)
df_cleaned['Pningkatn_Kpstas_Apartr_Desa_2022'] = df_cleaned['Pningkatn_Kpstas_Apartr_Desa_2022'].astype(int)
df_cleaned['Pningkatn_Kpstas_Apartr_Desa_2023'] = df_cleaned['Pningkatn_Kpstas_Apartr_Desa_2023'].astype(int)
df_cleaned['Pbrdayaan_Prempuan_2022'] = df_cleaned['Pbrdayaan_Prempuan_2022'].astype(int)
df_cleaned['Pbrdayaan_Prempuan_2023'] = df_cleaned['Pbrdayaan_Prempuan_2023'].astype(int)
df_cleaned['Koperasi_2022'] = df_cleaned['Koperasi_2022'].astype(int)
df_cleaned['Koperasi_2023'] = df_cleaned['Koperasi_2023'].astype(int)
df_cleaned['Dukgn_Pnanamn_Modal_2022'] = df_cleaned['Dukgn_Pnanamn_Modal_2022'].astype(int)
df_cleaned['Dukgn_Pnanamn_Modal_2023'] = df_cleaned['Dukgn_Pnanamn_Modal_2023'].astype(int)
df_cleaned['Pdagangan_Industri_2022'] = df_cleaned['Pdagangan_Industri_2022'].astype(int)
df_cleaned['Pdagangan_Industri_2023'] = df_cleaned['Pdagangan_Industri_2023'].astype(int)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
catVar = df_cleaned.select_dtypes(include = ['category'])
# Menghitung jumlah baris dan kolom subplot
num_variables = len(catVar.columns)
num_cols = 1  # Sesuaikan sesuai kebutuhan
num_rows = int(np.ceil(num_variables / num_cols))

# Membuat grid countplot dengan Seaborn
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(25, 10 * num_rows))

# Menggunakan Seaborn untuk membuat countplot dalam grid
for i, ax in enumerate(axes.flatten()):
    if i < num_variables:
        col = catVar.columns[i]
        top_10_values = catVar[col].value_counts().head(10).index
        sns.countplot(y=col, data=catVar[catVar[col].isin(top_10_values)], ax=ax,
                      order=top_10_values)
        ax.set_title(col)
    else:
        fig.delaxes(ax)

# Mengatur layout
plt.tight_layout()
plt.show()
```

```python
# Perbandingan Sumber Pendapatan Desa
sumber_pendapatan_columns = ['PAD_2023', 'DD_2023',
                              'Pajak_dan_Retribusi_2023',
                              'Al_DD_2023', 'Bantuan_Prov_2023',
                              'Bantuan_Kab_Kot_2023', 'Lainnya_2023',
                              'Total_APBDES_Tahun_2023']
sumber_pendapatan_data = df_cleaned[sumber_pendapatan_columns]
sumber_pendapatan_sorted = sumber_pendapatan_data.sum().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sumber_pendapatan_sorted.plot(kind='bar')
plt.title('Perbandingan Sumber Pendapatan Desa Tahun 2023')
plt.xlabel('Sumber Pendapatan Desa')
plt.ylabel('Total Pendapatan (dalam satuan tertentu)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Perbandingan Bidang Pemberdayaan Masyarakat Desa
bidang_pemberdayaan_columns = ['Klautn_PerIkanan_2023', 'Ptanian_Pternakn_2023',
                               'Pningkatn_Kpstas_Apartr_Desa_2023',
                               'Pbrdayaan_Prempuan_2023',
                               'Koperasi_2023',
                               'Dukgn_Pnanamn_Modal_2023', 'Pdagangan_Industri_2023',
                               'Pbrdayaan_Masyrkt_2023']
bidang_pemberdayaan_data = df_cleaned[bidang_pemberdayaan_columns]
bidang_pemberdayaan_sorted = bidang_pemberdayaan_data.sum().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
bidang_pemberdayaan_sorted.plot(kind='bar')
plt.title('Perbandingan Bidang Pemberdayaan Masyarakat Desa Tahun 2023')
plt.xlabel('Bidang Pemberdayaan Masyarakat Desa')
plt.ylabel('Total Anggaran (dalam satuan tertentu)')
plt.xticks(rotation=45)
plt.show()
```
