## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Data

```python
data = "/content/drive/Othercomputers/My Laptop/Magang MMB/Data1110 desa indonesia timur ~ Perbaikan.xlsx"
```

```python
df = pd.read_excel(open(data, 'rb'), sheet_name='geografi dan topografi desa', header=2)
df.drop(3, inplace=True)
df
```

```python
main = df.iloc[3:, :8]
main.head(5)
```

```python
geografi = df.iloc[:, 8:]
geografi.columns = geografi.iloc[1]
geografi = geografi.iloc[3:, :]
geografi.head(5)
```

```python
N, P = df.shape
print('df => baris = ', N, ', Kolom (jumlah variabel) = ', P)
N, P = main.shape
print('main => baris = ', N, ', Kolom (jumlah variabel) = ', P)
N, P = geografi.shape
print('geografi => baris = ', N, ', Kolom (jumlah variabel) = ', P)
```

```python
concat_df = pd.concat([main, geografi], axis=1)
concat_df.head(5)
```

```python
print("NAMA PROVINSI \n", concat_df["NAMA PROVINSI"].unique(), "\n")
print("NAMA KABUPATEN \n", concat_df["NAMA KABUPATEN"].unique(), "\n")
print("NAMA KECAMATAN \n", concat_df["NAMA KECAMATAN"].unique(), "\n")
print("Jenis Topografi Desa \n", concat_df["Jenis Topografi Desa"].unique(), "\n")
```

## Visualisasi

```python
# Assuming 'NAMA PROVINSI' contains unique province names
unique_provinsi = concat_df['NAMA PROVINSI'].unique()

# Initialize an empty list to store total luas for each province
total_luas_per_provinsi = []

# Calculate total luas for each province
for provinsi in unique_provinsi:
    total_luas = concat_df.loc[concat_df['NAMA PROVINSI'] == provinsi, 'Total Luas Wilayah Desa'].sum()
    total_luas_per_provinsi.append(total_luas)

# Create a bar plot
plt.figure(figsize=(15, 5))
bars = plt.bar(x=unique_provinsi, height=total_luas_per_provinsi)

# Add values on top of each bar
for bar, value in zip(bars, total_luas_per_provinsi):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.15, bar.get_height() + 0.05, f'{value:.2f}', ha='center', va='bottom')

# Add labels and title
plt.xlabel('Nama Provinsi')
plt.ylabel('Total Luas Wilayah Desa')
plt.title('Total Luas Wilayah Desa per Provinsi')

# Show the plot
plt.show()
```

```python
concat_df2 = concat_df.copy()

Q1 = concat_df2['Total Luas Wilayah Desa'].quantile(0.25)
Q3 = concat_df2['Total Luas Wilayah Desa'].quantile(0.75)
IQR = Q3 - Q1 #IQR is interquartile range.
print("Q1={}, Q3={}, IQR={}".format(Q1, Q3, IQR))
concat_df2.drop(concat_df2[(concat_df2['Total Luas Wilayah Desa'] > Q3+1.5*IQR)].index, inplace=True)
```


**Informasi yang di dapat dari table kawasan:**
*   data "Terdapat." adalah pilihan
*   data "Jumlah" adalah jumlah
*   data "lainnya" adalah isian



```python
columnsPil = kawasan.filter(like='a. Terdapat', axis=1)
for i in columnsPil.columns:
  print(i)
  print(columnsPil[i].unique())
```

```python
columnsPil = kawasan.filter(like='Terdapat', axis=1).columns
fig, axes = plt.subplots(nrows=len(columnsPil), ncols=1, figsize=(15, 3*len(columnsPil)))
for i, column_name in enumerate(columnsPil):
    x_label = concat_df[concat_df[column_name] == "Ada"]
    x = x_label['NAMA PROVINSI'].value_counts()
    axes[i].bar(x.index, x.values)
    axes[i].set_ylabel('Frekuensi')
    axes[i].set_title(column_name)

plt.tight_layout()
plt.show()
```
