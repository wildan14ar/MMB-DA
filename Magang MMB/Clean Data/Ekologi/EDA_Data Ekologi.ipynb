```python
# Membuat grid histogram dengan Seaborn
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=2, ncols=6, figsize=(30,15 ))

# Menggunakan Seaborn untuk membuat histogram dalam grid
for i in range(len(numVar.columns)):
    row, col = divmod(i,6)
    sns.histplot(numVar.iloc[:, i], kde=True, ax=axes[row, col], element="step", stat="count", common_norm=False)
    axes[row, col].set_title(numVar.columns[i])

# Mengatur layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
np.random.seed(42)
numVar = np.random.normal(loc=0, scale=1, size=100)

# Identify outliers using IQR method
Q1 = np.percentile(numVar, 25)
Q3 = np.percentile(numVar, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = (numVar < lower_bound) | (numVar > upper_bound)

# Create a bar plot for outliers
categories = ['No Outlier', 'Outlier']
outlier_counts = [len(numVar) - np.sum(outliers), np.sum(outliers)]

plt.bar(categories, outlier_counts, color=['blue', 'red'])
plt.title('Outliers in numVar')
plt.xlabel('Outlier Status')
plt.ylabel('Count')
plt.show()
```
